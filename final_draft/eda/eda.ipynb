```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_parquet('dataset_complete.parquet')
```

```python
gdp = df["GDP per capita (constant 2015 US$)"]
happiness = df["score"]

# original gdp vs happiness 
plt.figure(figsize=(7,5))
plt.scatter(gdp, happiness, alpha=0.6, color="steelblue")
plt.xlabel("GDP per capita (US$)")
plt.ylabel("Happiness Score")
plt.title("Happiness vs. GDP per capita (Non-linear)")
plt.tight_layout()
plt.show()


# log(gdp) vs happiness
gdp_log = np.log(gdp)

plt.figure(figsize=(7,5))
plt.scatter(gdp_log, happiness, alpha=0.6, color="darkorange")
plt.xlabel("log(GDP per capita)")
plt.ylabel("Happiness Score")
plt.title("Happiness vs. log(GDP per capita) (Linearized)")
plt.tight_layout()
plt.show()
```

```python
gni = df["GNI per capita (constant 2015 US$)"]

# original gni vs happiness 
plt.figure(figsize=(7,5))
plt.scatter(gni, happiness, alpha=0.6, color="steelblue")
plt.xlabel("GNI per capita (US$)")
plt.ylabel("Happiness Score")
plt.title("Happiness vs. GNI per capita (Non-linear)")
plt.tight_layout()
plt.show()


# log(gni) vs happiness
gni_log = np.log(gni)

plt.figure(figsize=(7,5))
plt.scatter(gdp_log, happiness, alpha=0.6, color="darkorange")
plt.xlabel("log(GNI per capita)")
plt.ylabel("Happiness Score")
plt.title("Happiness vs. log(GNI per capita) (Linearized)")
plt.tight_layout()
plt.show()
```
